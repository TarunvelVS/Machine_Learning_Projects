```python
def correlation_matrix(data):
    plt.figure(figsize=(15,6))
    heat_map = sns.heatmap(data.corr(), annot=True, cmap="BrBG")
    heat_map.set_title("Correlation Matrix")
    plt.show()
```

## Data Preprocessing

total_rooms, total_bedrooms, population, households are skewed to right or Positively skewed. Thus we need to do log transformation.

```python
data["total_rooms"] = np.log(data["total_rooms"]+1)
data["total_bedrooms"] = np.log(data["total_bedrooms"]+1)
data["population"] = np.log(data["population"]+1)
data["households"] = np.log(data["households"]+1)
```

```python
data.hist(figsize=(15,7))
```

Thus now the data looks like a Gusssian Bell Curve or Normal Distribution

Transforming the categorical value into numerical value for ocean proximity data column by means of creating dummies

```python
data.ocean_proximity.value_counts()
```

```python
ocean_proximity_dummies = pd.get_dummies(data.ocean_proximity)
```

```python
data = data.join(pd.get_dummies(data.ocean_proximity))
```

```python
data = data.drop(["ocean_proximity"],axis=1)
```

```python
data
```

```python
correlation_matrix(data)
```

## Feature Engineering and Analysis

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x="latitude", y="longitude", data=data, hue = "median_house_value", palette = "coolwarm")
```

```python
data["bedroom_ratio"] = data["total_bedrooms"]/data["total_rooms"]
data["household_rooms"] = data["total_rooms"]/data["households"]
```

## Linear Regression Model

```python
from sklearn.model_selection import train_test_split
```

```python
x = data.drop(["median_house_value"], axis = 1)
y = data["median_house_value"]

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
X_test
```

```python
reg.score(X_test, y_test)
```
